```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
#Clean Data
df = pd.read_csv('https://raw.githubusercontent.com/PythonicPastor/PokeWeight_Data/main/pokedex.csv?token=GHSAT0AAAAAABY4F3DPUV3OGXRMAEOJSKP2YZVQWNA')
df = df.drop(df.columns.difference(['name','height_m','weight_kg']), axis=1)

df['bmi'] = round(df['weight_kg']/(df['height_m']**2),1)
df = df[df['bmi']<100]
over = df[df['bmi']>25].sort_values('bmi')
under = df[df['bmi']<18.5].sort_values('bmi')

#Plot Data for Overweight Pokemon
plt.figure(figsize=(12, 250))
plt.margins(y=.0025)
plt.title('Overweight Pokemon', size=16)
plt.scatter(over['bmi'],over['name'],color='red')
plt.yticks(rotation = 45, size=16)

#Plot Data for Underweight Pokemon
plt.figure(figsize=(12, 85))
plt.margins(y=.0025)
plt.title('Underweight Pokemon',size=16)
plt.scatter(under['bmi'],under['name'],color='blue')
plt.yticks(rotation = 45, size=16)
plt.show()
```
